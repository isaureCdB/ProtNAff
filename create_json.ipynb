```python
import json

structures_json = json.load(open("structures.json"))
```

Instance of the filter given in the paper

It is extracting all single-stranded nucleotides in contact with the protein.

```python
# Define what is single-stranded
ss_set = set(["L", "T", "S", "J", "B", "I"])

for structure in structures_json.keys():
    pdb_info = structures_json[structure]
    for chain_id in pdb_info["nachains"]:
        chain_id = "chain_" + chain_id
        try:
            # Extract nucleotide in contact with the protein
            interf_prot = pdb_info["interface_protein"]["model_1"][chain_id]
            nucl_interf = [pdb_info["mapping"][chain_id][element.split("_")[1]] \
                           for element in list(interf_prot.keys())]
            # List of secondary structures for nucleotides
            ss = pdb_info['ss'][chain_id]
            # List of single-stranded nucleotides
            nucl_ss = [pdb_info["mapping"][chain_id][n.split("_")[1]] \
                       for n in ss if ss[n][0] in ss_set]
            # List of single-stranded nucleotides in contact with the protein
            nucl_interf_ss = [int(n) for n in nucl_interf if n in nucl_ss]
            print("PDB id : {} \n RNA chain : {}\n Nucl in contact and ss : {}".format(structure, \
                                                                     chain_id, nucl_interf_ss))
        except:
            pass
```

First statistics on X-ray and RMN structures

```python
result = {"x-ray":{"ss":0, "ds":0}, "nmr":{"ss":0, "ds":0}}

x = 0
n = 0

for value in structures_json.values():
    if value["method"] == "x-ray diffraction":
        x += 1
        for chain in value['ss'].values():
            for res in chain.values():
                if res[0] == "D":
                    result["x-ray"]["ds"] += 1
                else:
                    result["x-ray"]["ss"] += 1
    elif value["method"] == "solution nmr" or value["method"] == "solution nmr; theoretical model" or value["method"] == "solid-state nmr":
        n += 1
        for chain in value['ss'].values():
            for res in chain.values():
                if res[0] == "D":
                    result["nmr"]["ds"] += 1
                else:
                    result["nmr"]["ss"] += 1
print(result)
x_per = round((result["x-ray"]['ss']*100/(result["x-ray"]['ss'] + result["x-ray"]['ds'])),2)
rmn_per = round((result["nmr"]['ss']*100/(result["nmr"]['ss']+result["nmr"]['ds'])),2)
print("Percentages ss: X-ray: {}%, NMR: {}%".format(x_per, rmn_per))
```

Plot for clustering modularity (You need to run the create_frag_library.sh first !)

```python
import matplotlib.pyplot as plt
import numpy as np

def count_cluster_size(motif):
    clust1 = np.load("./trilib/{}-dr0.2r-clust1.0.npy".format(motif)).shape[0]
    clust3 = np.load("./trilib/{}-dr0.2r-clust1.0-clust3.0.npy".format(motif)).shape[0]
    return clust1, clust3

legend = ['AAA', 'AAC', 'ACA', 'ACC', 'CAA', 'CAC', 'CCA', 'CCC']
clust1 = []
clust3 = []
for motif in legend:
    motif1, motif3 = count_cluster_size(motif)
    clust1.append(motif1)
    clust3.append(motif3)

plt.bar([1, 2, 3, 4, 5, 6, 7, 8], clust1, 0.4, label="cluster 1A", log=True, color='green')
plt.bar([1.4, 2.4, 3.4, 4.4, 5.4, 6.4, 7.4, 8.4], clust3, 0.4, label="cluster 3A", log=True, color='orange')
plt.xticks([1.2, 2.2, 3.2, 4.2, 5.2, 6.2, 7.2, 8.2], legend)
plt.ylabel("Number of clusters (logarithm scale)")
plt.legend()
plt.show()
```
